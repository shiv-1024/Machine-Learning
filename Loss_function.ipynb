```python
import numpy as np
import pandas as pd

def bce(y,y_pred):
    loss = -(y* np.log(y_pred) + (1-y)*np.log(1-y_pred))
    
    return loss


y = 0
y_pred = 0.9

loss_fun = bce(y,y_pred)

print(loss_fun)
```

```python
def cce(y,y_pred):
    loss = -np.sum(y*np.log(y_pred))

    return loss


y = [0,1,0]
y_pred = [0.1,0.7,0.2]

cce = cce(y,y_pred)
print(cce)
```

```python
def scce(y,y_pred):
    loss = -(np.log(y_pred[y]))

    return loss


y = 1
y_pred = [0.1,0.7,0.2]

scce = scce(y,y_pred)
print(scce)
```
